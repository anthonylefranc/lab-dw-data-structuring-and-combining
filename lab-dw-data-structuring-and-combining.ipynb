Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
#Importing libraries
import pandas as pd
import numpy as np

#Importing datafiles
file_1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
file_2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2_clean.csv')
file_3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3_clean.csv')

#verifying that columns label are the same for each datasets
#file_1.columns == file_2.columns
#file_2.columns == file_3.columns
#file_1.columns == file_3.columns

#Concatenating the three files to get a dirty version of customer_data

customer_data_dirty = pd.concat([file_1, file_2, file_3], axis=0)

#cleaning columns labels

customer_data_dirty = customer_data_dirty.rename(columns={'Customer':'customer',
                                'ST':'state',
                                'GENDER':'gender',
                                'Customer Lifetime Value':'customer_lifetime_value',
                                'Income':'income',
                                'Monthly Premium Auto':'monthly_premium',
                                'Number of Open Complaints':'open_complaints',
                                'Policy Type':'policy_type','Vehicle Class':'vehicule_class','Total CLaim Amount':'total_claim_amount'})


#check datatypes
#print(customer_data_dirty.dtypes)
#open complaints & CLV might be numerical to check

#check Null values >>> 8811 NaN for each labels except for Gender who has 9172
#ustomer_data_dirty.isna().sum()
#ustomer_data_dirty.shape

#checking which rows might have null valueand removing them
#customer_data_dirty[customer_data_dirty['customer_lifetime_value'].isna()==True]
#customer_data_dirty.isna().sum()

customer_data_dirty.dropna(subset=['state'], inplace=True)
customer_data_dirty.dropna(subset=['customer_lifetime_value'], inplace=True)
customer_data_dirty.dropna(subset=['gender'], inplace=True)

#checking for naming errors in gender & customer_lifetime_value(feamale, or percentage in CLV) 

customer_data_dirty['gender'].replace({'Femal': 'F', 'Male': 'M', 'female': 'F'}, inplace=True)
customer_data_dirty['state'].replace({'Arizona': 'AZ', 'Nevada': 'NV', 'California': 'CA', 'Cali': 'CA', 'Washington': 'WS', 'Oregon': 'OR'}, inplace=True)
customer_data_dirty['vehicule_class'].replace({'SUV': 'Luxury', 'Luxury Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Sports Car': 'Luxury'}, inplace=True)
customer_data_dirty['customer_lifetime_value'] = customer_data_dirty['customer_lifetime_value'].str.replace('%', '').astype(float)
customer_data = customer_data_dirty['open_complaints'] = customer_data_dirty['open_complaints'].str.split('/').str[0]

#customer_data_dirty['vehicule_class'].unique()
display(customer_data)


```

```python


#removing the empty unnamed:0 column 
#marketing_data = marketing_data.drop(['unnamed:_0'], axis=1)

marketing_data['customer_lifetime_value'] = round(marketing_data['customer_lifetime_value'], 2)

#creating the pivot table (i don't know from context if customer value is equal to amount of sales, but that's my assumption here)
marketing_summary = pd.pivot_table(marketing_data, values='customer_lifetime_value', index='sales_channel', aggfunc='sum')

# Rounding down to 2 decimals
marketing_summary = marketing_summary.round(2)

display(marketing_summary)
```
